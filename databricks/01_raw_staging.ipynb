# Raw Layer

### The Handshake

```python
# We retrieve the key safely without printing it
storage_key = dbutils.secrets.get(scope="kv-scope", key="ADLS-Storage-Key")
storage_name = "aerostreamvariance" 

# We tell Spark: "Whenever you see this storage account, use this key"
spark.conf.set(
    f"fs.azure.account.key.{storage_name}.dfs.core.windows.net",
    storage_key
)

print(f"Successfully connected to {storage_name}!")
```

Creating raw schema that holds data in tables after parsing from json

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS raw")
```

# Flights

```python
#Read the JSON from the Bronze container
df_json = spark.read.json(f"abfss://raw@{storage_name}.dfs.core.windows.net/flights/")
```

```python
# Write it to a Delta Table
(df_json.write
    .format("delta")
    .mode("append")
    .option("mergeSchema", "true")
    .saveAsTable("raw.flights_bronze")
)

print("Data successfully loaded into table 'raw.flights_raw'")
```

```python
# verifying the data
display(spark.sql("SELECT * FROM raw.flights_bronze"))
```

# Weather

```python
weather_path = f"abfss://raw@{storage_name}.dfs.core.windows.net/weather/"
```

```python
df_weather_raw = spark.read.option("multiline", "true").json(weather_path)
```

```python
print("Inferred Schema:")
df_weather_raw.printSchema()
```

### Data Preview

```python
print("\nData Preview:")
display(df_weather_raw)
```

```python
(df_weather_raw.write
    .format("delta")
    .mode("append") # append so history builds up over time
    .option("mergeSchema", "true") # In case OpenMeteo adds new fields later
    .saveAsTable("raw.weather_bronze")
)

print("Data successfully loaded into table 'raw.weather_bronze'")
```
